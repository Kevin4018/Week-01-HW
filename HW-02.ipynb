### **Extension 1: Using a Sequence of Previous Words (n-grams)**

Instead of just focusing on the last word, we can use the last *n* words as context to decide the next word. This method is called an **n-gram model**. By using more words in the history, the chatbot can make more informed predictions about the next word, resulting in better responses.

Here’s how the code can be extended to consider the last two words (a bi-gram) instead of just one:

```python
# Initialize the dictionaries for counting word sequences and transitions
word_used = dict()  # To count how often each sequence of words appears
next_word = dict()  # To store the frequencies of the next words based on a sequence

n = 2  # We are using bi-grams, so we track the last 2 words

# Loop over words, but now track sequences of n previous words (n-grams)
for i in range(len(words) - n):
    # Create the sequence (n-gram) of the last `n` words
    current_sequence = tuple(words[i:i+n])  # e.g., ('the', 'cat')

    # Track how often this sequence appears
    if current_sequence in word_used:
        word_used[current_sequence] += 1
    else:
        word_used[current_sequence] = 1
        next_word[current_sequence] = {}

    # Look at the next word following this sequence
    next_word_in_sequence = words[i + n]

    # Track how often the next word follows the current sequence
    if next_word_in_sequence in next_word[current_sequence]:
        next_word[current_sequence][next_word_in_sequence] += 1
    else:
        next_word[current_sequence][next_word_in_sequence] = 1
```

#### **What Changed:**
- We are now tracking **bi-grams** instead of just single words. This means the chatbot uses the last two words to predict the next word.
- `current_sequence` is a tuple of the last two words (like `('the', 'cat')`).
- This results in better predictions since it looks at more context instead of just the immediate previous word.

### **Extension 2: Visualizing the Markov Chain Structure (with a Diagram)**

A useful extension is to visualize the Markov Chain structure as a **directed graph**. Each word (or sequence of words) is a node, and an arrow from one word to another represents a possible transition, with the thickness of the arrow proportional to the frequency of that transition.

You could use a library like `networkx` and `matplotlib` in Python to generate such a diagram. Here's how the visualization code might look:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges based on the word transitions
for sequence, transitions in next_word.items():
    for next_word_in_sequence, count in transitions.items():
        # Add an edge from the sequence to the next word with the weight as the count
        G.add_edge(' '.join(sequence), next_word_in_sequence, weight=count)

# Draw the graph with weights (frequencies)
pos = nx.spring_layout(G)  # Position nodes using a force-directed algorithm
edges = G.edges(data=True)

# Draw nodes and labels
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold")
# Draw edges with their weights
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in edges})

# Display the graph
plt.show()
```

#### **Explanation of Visualization:**
- This graph shows how words (or word sequences) transition to other words.
- Each node represents a word or sequence of words.
- Each arrow (directed edge) represents a transition from one word/sequence to another.
- The thickness or label on the arrows represents the frequency of each transition, giving you insight into how words are likely to follow each other.

This diagram helps visualize the **Markov Chain** structure, making it easier to understand how the chatbot is predicting the next word.
